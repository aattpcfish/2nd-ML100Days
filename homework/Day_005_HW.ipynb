```python
Sel_col = app_train.select_dtypes(include=["int64","float64"]).columns
```

## 所有非object 的欄位的統計值

```python
print(app_train[Sel_col].agg(['mean','std','min','max']).T)
```

## 所有非object 的欄位的直方圖

```python

fig, axs = plt.subplots(12, 3, constrained_layout=True,figsize=(14, 30))

for i,col in enumerate(app_train[Sel_col].columns[0:36]):
    axs[i/3][i%3].hist(app_train[col].dropna())
    axs[i/3][i%3].set_title(col)
    
    for tick in axs[i/3][i%3].get_xticklabels():
        tick.set_rotation(45)
    
plt.show()
```
